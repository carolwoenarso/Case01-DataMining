```python
# Jakarta
dfJ_cat = dfJ.select_dtypes(include = ['category'])

for col in dfJ.columns:
  print(col,': ',set(dfJ[col].unique()))

print(dfJ.isnull().sum())
```

```python
dfJ2 = dfJ.drop(columns=['Unnamed: 14', 'Unnamed: 15', 'Unnamed: 16', 'Unnamed: 17', 
                  'Unnamed: 18', 'Unnamed: 19', 'Unnamed: 20', 'Unnamed: 21', 
                  'Unnamed: 22', 'Unnamed: 23', 'Unnamed: 24', 'Unnamed: 25'])
dfJ3 = dfJ2.dropna(how='all')
dfJ3
```

```python
dfJ3.shape
```

```python
for col in dfJ3.columns:
  print(col,': ',set(dfJ3[col].unique()))

print(dfJ3.isnull().sum())
```

```python
dfJ4 = dfJ3.drop(columns=['garasi','hadap'])
dfJ4['harga']=dfJ4['harga']//1000000
dfJ4
```

```python
numericalJ = dfJ4.select_dtypes(include = ['int64','float64']).columns

dfJ4[numericalJ].hist(figsize=(15, 6), layout=(2, 4));
```

```python
dfJ5 = dfJ4.fillna(dfJ4.median())
dfJ6 = dfJ5.fillna(dfJ5.mode())
print(dfJ6.isnull().sum())
```
